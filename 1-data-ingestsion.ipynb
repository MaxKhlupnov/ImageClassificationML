```python
import os
from os import path
from glob import glob, iglob
from PIL import Image
import matplotlib.pyplot as plt

save_path = 'C:\\PROJECTS\\DATA\\Train\\'

files_path = path.join(save_path, 'bus')
fail_files = sorted(iglob(path.join(files_path, 'bus*.jpg')))

# import matplotlib.pyplot as plt
pil_image = Image.open(fail_files[0])
print(pil_image.size)
pil_image

width = [Image.open(file).size[0] for file in fail_files]
height = pil_image.size[1]

plt.hist(width, bins=10)
plt.title('Image Width Distribution')
plt.show()
```
